```python
import pandas as pd

# Carga el archivo CSV
df = pd.read_csv('avocado.csv')

# Muestra las primeras filas del DataFrame
df.head()


df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month


resultados = df.groupby(['Month', 'type'])['Total Bags'].sum().reset_index()
max_bags = resultados['Total Bags'].max()
mes_estilo_maximo = resultados[resultados['Total Bags'] == max_bags]

print("Mes y estilo con la mayor cantidad de bolsas:")
print(mes_estilo_maximo)


```

```python
import pandas as pd

# Lee el archivo .csv
df = pd.read_csv('avocado.csv')

nuevo_df = df[['Date', 'AveragePrice', 'Total Volume']]
print("Datos remarcables ")
print(nuevo_df.head())

# Revisa los datos
descripcion = df.describe()
print("Descripción estadística de los datos:")
print(descripcion)


```

```python
# Calcula la media, moda y mediana de todos los datos numéricos relevantes
media = df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']].mean()
moda = df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']].mode().iloc[0]
mediana = df[['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']].median()

print("Media de los datos numéricos relevantes:")
print(media)

print("\nModa de los datos numéricos relevantes:")
print(moda)

print("\nMediana de los datos numéricos relevantes:")
print(mediana)
```
